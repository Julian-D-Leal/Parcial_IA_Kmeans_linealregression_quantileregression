```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
#Solución punto 2 parcial ia, con dataset de Cancer Linear Regression Model Tutorial con el archivo death.cvs con muertes por cancer

data = pd.read_csv("death.csv")
dates = data['County'] # Guardamos la columna Date.
data = data.drop('County',1) # Borramos la columna del dataframe.

data
```

```python
# Revisar nulos
print(data.info())

# Revisar atipicos
print(data.describe())
```

```python
#se reemplaza los datos categoricos a numericos
varlist =  ['RecentTrend2']

def binary_map(x):
    return x.map({'falling': 1, "stable": 0})

# # Aplicar la función a cada fila del dataset
data[varlist] = data[varlist].apply(binary_map)
```

```python
#se reemplaza los datos categoricos a numericos
varlist_2 =  ['Metobjective45p']

def binary_map_2(x):
   return x.map({'No': 0, "Yes": 1})

data[varlist_2] = data[varlist_2].apply(binary_map_2)
```

```python
data
```

```python
#se reemplazan los asteriscos por datos nulos
data[data == '*'] = np.NAN
data[data == '**'] = np.NAN
data
```

```python
#se eliminaron todas las filas con datos nulos que eran asteriscos
data.dropna(inplace=True)
data
```

```python
#se cambian los datos a tipos flotantes y enteros dependiendo el caso
data['Age-AdjustedDeathRate'] = data['Age-AdjustedDeathRate'].astype(float, errors = 'raise')
data['Lower95%forDeathRate'] = data['Lower95%forDeathRate'].astype(float, errors = 'raise')
data['Upper95%forDeathRate'] = data['Upper95%forDeathRate'].astype(float, errors = 'raise')
data['AvgDeathsperYear'] = data['AvgDeathsperYear'].astype(int, errors = 'raise')
data['Recent5YearTrend2inDeathRates'] = data['Recent5YearTrend2inDeathRates'].astype(float, errors = 'raise')
data['Lower95%forTrend'] = data['Lower95%forTrend'].astype(float, errors = 'raise')
data['Upper95%forTrend'] = data['Upper95%forTrend'].astype(float, errors = 'raise')
data.info()
```

```python
#normalizamos los datos para que esten en la misma escala
scaler = MinMaxScaler()

# Se aplica a todas las columnas menos a las dummies y a las si/no que discutimos arriba
num_vars = ['FIPS', 'Age-AdjustedDeathRate', 'Lower95%forDeathRate', 'Upper95%forDeathRate', 'AvgDeathsperYear','Recent5YearTrend2inDeathRates','Lower95%forTrend','Upper95%forTrend']
data_scaler = scaler.fit_transform(data[num_vars])

data_scaler
```

```python
#se ejecuta el metodo del codo para obtener el número de clusters
kmeans_kwargs = {
"init": "k-means++",
"n_init": 50,
"max_iter": 500
}

# A list holds the SSE values for each k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(data_scaler)
    sse.append(kmeans.inertia_)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
from kneed import KneeLocator

#nos retorna el número de clusters donde se obtiene menor diferencia con los demás puntos
kl = KneeLocator(
range(1, 11), sse, curve="convex", direction="decreasing"
)

kl.elbow
```

```python
#Desde aquí se hace k-means y se usaran 3 clusters

kmeans = KMeans(
init="k-means++",
n_clusters=3,
n_init=50,
max_iter=500,
)
```

```python
kmeans.fit(data_scaler)
```

```python
# The lowest SSE value
kmeans.inertia_
```

```python
# Final locations of the centroid
kmeans.cluster_centers_
```

```python
# The number of iterations required to converge
kmeans.n_iter_
```

```python
#agregamos a que cluster pertenece cada una de las filas en una nueva columna llamada label

labels = kmeans.predict(data_scaler)
data['label'] = labels
```

```python
#Se agrega cada estado a su respectiva fila a la tabla original
data.insert(0, 'County', dates)
```
